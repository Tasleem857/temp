#### PreProcessing

```python
#Splitting the data
X = train_df.drop(columns=['Sales', '#Order', 'Date'])
y = train_df['Sales']
```

```python
X_test = test_df.drop(columns=['Date'])
```

```python
X.head()
```

```python
#Required modules for the pipeline
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
categorical_features = ['Store_Type', 'Location_Type', 'Region_Code', 'Discount']
```

```python
#Pipeline for transformation
categorical_transformer = Pipeline(steps=[
    ('ordinal', OrdinalEncoder())
])
```

```python
#Pipeline for column transformation
preprocessor = ColumnTransformer(transformers=[
    ('categorical', categorical_transformer, categorical_features)
])
```

```python
from xgboost import XGBRegressor

xgb = XGBRegressor()
```

```python
pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('XGBRegressor', xgb)])
```

```python
pipeline.fit(X,y)
```

```python
y_pred = pipeline.predict(X)
```

```python
from sklearn.metrics import mean_squared_log_error as lmse
lmse(y, y_pred)
```

```python
y_new = pipeline.predict(X_test)
```

```python
sales = pd.DataFrame(y_new, columns=['Sales'])
left = pd.DataFrame(X_test.index)
final = pd.concat([left, sales], axis=1)
```

```python
final.to_csv('Final.csv', index=False)
```
